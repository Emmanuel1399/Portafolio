```python
import torch as torch

import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

#Vectores normales w1 y w2
w1 = torch.tensor([0.5, 0.2])
w2 = torch.tensor([-0.1, 0.05])

#create 1D tensors
step = 0.01
x_1 = torch.arange(-1.0, 5.0, step)
x_2 = torch.arange(-1.0, 6.0, step)
#Create 2D tensors with variation along the two axis
X_1, X_2 = torch.meshgrid(x_1, x_2)

# funciones f1 y f2
f1 = w1[0] * X_1 + w1[1] * X_2
f2 = w2[0] * X_1 + w2[1] * X_2

# Graficar plano f1
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_1.numpy(), X_2.numpy(), f1.numpy())
plt.show()

# Graficar plano f2
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_1.numpy(), X_2.numpy(), f2.numpy())
plt.show()


# Curva de nivel f1
plt.title('Curva de Nivel F1 (w1)')
plt.xlabel('X_1')
plt.ylabel('X_2')
contours = plt.contour(X_1.numpy(), X_2.numpy(), f1.numpy())
#Vector
plt.quiver(1, 1, w1[0], w1[1], angles='xy', scale_units='xy', scale=1)
plt.clabel(contours, inline=1, fontsize=10)
plt.show()

# Curva de nivel f2
plt.title('Curva de Nivel F2 (w2)')
plt.xlabel('X_1')
plt.ylabel('X_2')
contours = plt.contour(X_1.numpy(), X_2.numpy(), f2.numpy())
plt.quiver(1, 1, w2[0], w2[1], angles='xy', scale_units='xy', scale=1)
plt.clabel(contours, inline=1, fontsize=10)
plt.show()

```
